```python
import pylab
from mpl_toolkits.mplot3d import Axes3D
import numpy as np
```

```python
x = np.arange (-10, 10, 0.1)
y = np.arange (-10, 10, 0.1)
x, y = np.meshgrid(x, y)

U = y*x**2 + (y**3)/3 + 2*x**2 + 3*y**2 - 1 

fig = pylab.figure()
axes = Axes3D(fig)

axes.plot_surface(x, y, U)

pylab.show()
```
