1. Задание (в программе)
Нарисуйте график функции:
y(x) = k∙cos(x – a) + b
для некоторых (2-3 различных) значений параметров k, a, b

```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
x = np.linspace(-4 * np.pi, 4 * np.pi, 1000)
k = np.array([2., 4., 8.])
a = np.array([np.pi / 4, np.pi / 3, np.pi / 2])
b = [2.,3., 4.]
plt.xlabel("x")
plt.ylabel("y")
plt.xlim(-10, 10)
for i in range(3):
    plt.plot(x, k[i] * np.cos(x - a[i]) - b[i])
plt.grid(True)
```

3. Задание (в программе)
Напишите код, который будет переводить полярные координаты в декартовы.
Напишите код, который будет рисовать график окружности в полярных координатах.
Напишите код, который будет рисовать график отрезка прямой линии в полярных координатах.

```python
def to_decart(r, alpha):
    x = r * np.cos(alpha)
    y = r * np.sin(alpha)
    return x, y

x, y = to_decart(1, np.pi + np.pi / 4)
print(x, y)
```

Напишите код, который будет рисовать график отрезка прямой линии в полярных координатах.

```python
def to_polar(x, y):
    r = np.sqrt(x**2 + y**2)
    if (x == 0) and (y == 0):
        alpha = 0
    elif (x == 0) and (y > 0):
        alpha = np.pi / 2
    elif (x == 0) and (y < 0):
        alpha = 3 * np.pi / 2    
    elif (x > 0) and (y == 0):
        alpha = 0
    elif (x < 0) and (y == 0):
        alpha = np.pi        
    elif (x < 0) and (y > 0):
        alpha = np.pi - np.arcsin(y / r)
    elif (x < 0) and (y < 0):
        alpha = 2 * np.pi - np.arccos(x / r)
    elif (x > 0) and (y < 0): 
        alpha =  -np.arccos(x / r)
    else :   
        alpha = np.arctan(y / x)
    return r, alpha

r, alpha = to_polar(-3, -3)
print(r, alpha)
```

```python
fig = plt.figure()
size = 100
x = np.linspace(-5, 5, size)
y =  -2 * x + 3
r = []
t = []
for i in range(size):
    r_, t_ = to_polar(x[i], y[i])
    r.append(r_)
    t.append(t_)
plt.polar(t, r, color="green" ,linewidth=3)
plt.grid(True)
```

4. Задание (в программе)
Решите систему уравнений:
y = x**2 – 1
exp(x) + x∙(1 – y) = 1

```python
from scipy.optimize import fsolve
```

```python
x = np.linspace(-3, 3, 201)
plt.plot(x, x**2 - 1) 
plt.plot(x, (np.exp(x) + x - 1) / x) 
plt.grid(True)
```

```python
def func(p):
    x, y = p
    return (x**2 - y - 1, np.exp(x) + x - x*y - 1)

x1, y1 = fsolve(func, (2.8, 7))
print(x1, y1)

x2, y2 = fsolve(func, (-2, 2))
print(x2, y2)
```
