3. Даны три множества a,b и с. Необходимо выполнить все изученные виды бинарных операций над всеми комбинациями множеств.
*Выполнить задание 1 на языке Python

```python
import itertools
a = set([1, 2, 3, 4, 5, 6])
b = set([2, 4, 6, 7, 9, 11])
c = set([1, 3, 5, 8, 10, 12])

def decart(s1, s2):
    s_decart = set()
    for e in itertools.product(s1, s2):
        s_decart.add(e)
    return s_decart

print("Пересечение a b")
print(a.intersection(b))
print("Объединение a c")
print(a.union(c))
print("Разность c a")
print(c.difference(a))
print("Симметрическая разность c a")
print(c.symmetric_difference(a))
print("Декартово произведение b c")
sd = decart(b, c)
print(sd)
print(len(sd))
```

(-1)^2n + 1 / (n^2)     - ограничена


```python
# пятый член

import math
n = 5
print(1, 2 ** n - n)
print(2, 1 / (1 - (n + 1)))
print(3, (-1) ** n + math.sqrt(2 * n))
print(4, (-1) ** (2 * n) + 1 / (n**2))
```

*На языке Python предложить алгоритм вычисляющий численно предел с точностью 
10^-7
n / n!^(1/n)

```python
eps = 0.0000001

def f1(n):
    if n == 1: return 1
    z = 1
    for i in range(2, n + 1):
        z *= i ** (1 / n)
    return n / z

n = 2
el_n = f1(n)
el_n1 = f1(n + 1)
while el_n1 - el_n >= eps:
    n += 1
    el_n, el_n1 = el_n1, f1(n + 1)
else:
    print("Величина предела с точностью 10^-7")
    print(el_n1)
```
