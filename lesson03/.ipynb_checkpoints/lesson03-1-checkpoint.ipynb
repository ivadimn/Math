3. Задание (в программе)
Напишите код на Python, реализующий построение графиков:
окружности,
эллипса,
гиперболы.

```python
rcParams["figure.figsize"] = 4, 4
t = np.linspace(0, 2*np.pi, 101)
x = 3 * np.sin(t)
y = 3 * np.cos(t)
plt.xlim(-4, 4)
plt.ylim(-4, 4)
plt.plot(x,y)
plt.xlabel("x")
plt.ylabel("y")
```

```python
x = 3.5 * np.sin(t)
y = 2 * np.cos(t)
plt.xlim(-4, 4)
plt.ylim(-4, 4)
plt.plot(x,y)
plt.xlabel("x")
plt.ylabel("y")
```

гипербола 5x**2 - 4y**2 = 20
5x**2 - 4y**2 - 20 = 0
другая гипербола x**2 - y**2 - 1 = 0

```python
t = np.linspace(-4, 4, 101)
x, y = np.meshgrid(t, t)
plt.contour(x, y, 5*(x**2) - 4*(y**2) -20, [0], linewidths=4, colors="green")
plt.contour(x, y, (x**2) - (y**2) - 1, [0], linewidths=2, colors="red")
plt.xlim(-4, 4)
plt.ylim(-4, 4)
plt.xlabel("x")
plt.ylabel("y") 
plt.grid(True)
```

5. Задание (в программе)
Нарисуйте трехмерный график двух параллельных плоскостей.
Нарисуйте трехмерный график двух любых поверхностей второго порядка.

Плоскости
x + 2y + z + 2 = 0, 
2x + 4y + 2z + 10 = 0

```python
fig = plt.figure(figsize=(8, 6))
ax = Axes3D(fig)
X = np.arange(-5, 5, 1)
Y = np.arange(-5, 5, 1)
X, Y = np.meshgrid(X, Y)
Z = X + 2*Y + 2
Z2 = (2*X + 4*Y) / 2 + 10
ax.plot_surface(X, Y, Z)
ax.plot_surface(X, Y, Z2)
show()
```

Поверхности: 
шар
гиперболойд

```python
fig = plt.figure(figsize=(8, 6))
ax = Axes3D(fig)
u = np.linspace(-2 * np.pi, 2 * np.pi, 51)
v = np.linspace(-np.pi / 2, np.pi / 2, 51)
x = np.outer(np.cos(u), np.cos(v))
y = np.outer(np.sin(u), np.cos(v))
z = np.outer(np.ones(np.size(u)), np.sin(v))
ax.plot_surface(x, y, z, rstride=2, cstride=4, color="0.75")
show()
```

```python
fig = plt.figure(figsize=(8, 6))
ax = Axes3D(fig)
t = np.linspace(-4, 4, 101)
x, y = np.meshgrid(t, t)
z = np.sqrt(2*x**2 + 5*y**2 - 1)
ax.plot_surface(x, y, z, rstride=1, cstride=1, color="0.75")
ax.plot_surface(x, y, -z, rstride=1, cstride=1, color="0.75")
show()
```
